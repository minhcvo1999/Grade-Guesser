# Abstract

We created this project with the intention of answering the question: Will the grade you receive from a certain course be dependent on the quarter you take that class? We hoped to create some sort of model that could take all these independent variables into consideration to create an accurate model that would allow students to plan their schedules more efficiently. This tool would be useful for making adjustments and overall assisting in the challenge of planning your upcoming quarters and schedule. We aimed to use a variety of resources, such as professor ratings, their income, average grade, and previous yearly and quarterly trends, to try to find a correlation in order to create a prediction. We used the UCSD Capes dataset along with the UC Income Dataset and merged them into one dataset, which allowed us to create more plots with different features. We used a random forest regression tree to create a prediction model with 100 estimators to optimize efficiency, given the large quantity of data that needs to be analyzed (and multiple feature variables). After training the model on 80% of our data and testing it on 20%, our model came up inconclusive and inaccurate. We were only able to account for about 40% of the variance with an even smaller amount for unseen data. We ran a correlation matrix and determined from the results that the feature variables we hoped to use to create a prediction model had little to no correlation with each other. Hence, there is a lot of variance that we couldn't account for. We also used hypothesis testing to determine if our null hypothesis is valid or not: the GPA you get for taking a certain course is not affected by which quarter you are taking the course. We ended up using a t-test to compare fall, winter, and spring quarters for each year and determined that the majority of t-tests stated that we should fail to reject the null hypothesis, giving validity that our null hypothesis is mostly true.

### Cleaning the Data

```python
#List of function used to standardize the data

#Function used to clean Average Grade Expected and Average Grade Received columns
def standardize_grade(string):

    if isinstance(string, str):
        string = string.strip()

        string = string.replace('(', '')
        string = string.replace('A', '')
        string = string.replace('B', '')
        string = string.replace('C', '')
        string = string.replace('D', '')
        string = string.replace('F', '')
        string = string.replace('+', '')
        string = string.replace('-', '')
        string = string.replace('(', '')
        string = string.replace(')', '')

        string = string.strip()
        string = float(string)


    return string

#Function used to clean Percentage Recommended Class and Percentage Recommended Professor columns
def standardize_percentage(string):
    if isinstance(string, str):
        string = string.strip()

        string = string.replace('%', '')
        string = float(string)

    return string

#Function used to clean Course column
def standardize_course(string):
    value = string.split('-')
    string = value[0].strip()
    return string

#Function used to extract a Department name
def standardize_department(string):
    value = string.split(' ')
    string = value[0].strip()
    return string
```

```python
#Using function standardize_grade to drop the letter grade and keep the GPA of course
df['Average Grade Expected'] = df['Average Grade Expected'].apply(standardize_grade)
df['Average Grade Received'] = df['Average Grade Received'].apply(standardize_grade)

#Using function standardize_percentage to change the convert column from string to float
df['Percentage Recommended Class'] = df['Percentage Recommended Class'].apply(standardize_percentage)
df['Percentage Recommended Professor'] = df['Percentage Recommended Professor'].apply(standardize_percentage)

#Using function standardize_course to retrieve only the Course name and number
df['Course'] = df['Course'].apply(standardize_course)

#Using function standardize_department to create another column of just the department
department = df['Course'].apply(standardize_department)
df['Department'] = department

df
```

### Analyze the missingness in Average_Grade_Received

#### Analyze if Average_Grade_Received is MAR FROM "Department" column¶

```python
#Create a new column of bool dtype where Average_Grade_Received is missing is True, otherwise False
df = df.assign(Average_Grade_Received_missing = df['Average_Grade_Received'].isna())
```

```python
#Computing a observed statistic
pivoted = (
    df
    .pivot_table(index = 'Department', columns = 'Average_Grade_Received_missing', aggfunc='size')
    .apply(lambda x: x/x.sum())
    ).fillna(0)

obs_tvd = pivoted.diff(axis=1).iloc[:, -1].abs().sum() / 2
obs_tvd
```

```python
#Performing Permutation testing to see if the missingness is MAR
#We are computing tvd, basically a difference in distribution when we shuffle the Department column
#We are doing this shuffling 500 times and compute the result of each tvd in tvds
n_repetitions = 500
shuffled = df.copy()

tvds = []
for _ in range(n_repetitions):
    shuffled['Department'] = np.random.permutation(shuffled["Department"])

    pivoted = (
        shuffled
        .pivot_table(index = 'Department', columns = 'Average_Grade_Received_missing', aggfunc='size')
        .apply(lambda x: x/x.sum())
    )

    tvd = pivoted.diff(axis=1).iloc[:, -1].abs().sum() / 2
    tvds.append(tvd)
```

```python
#Compute the p-value
np.mean(np.array(tvds) >= obs_tvd)
# Conclusion: We reject the null and claim that missingness in Average_Grade_Received are depended on Department!
```

#### Because Average_Grade_Received is MAR, we will want to perform imputation on this, but there is one problem

As you can see, there is some value in this pivot table where the value is 0. This means that there is no class where there was Average_Grade_Received was recorded in that department

##### Problem: Imputation is physically impossible for some departments since some department doesn't record Average_Grade_Received AT ALL

##### IDEA: We can impute the values in expected average grades instead

To do this, we need to see if the distribution of  Average_Grade_Expected and Average_Grade_Received are similar
- We will find the average difference between expected and actual grades to see if there is significant difference between these distribution

```python
#Positive number = Average_Grade_Received > Average_Grade_Expected
#Negative number = Average_Grade_Expected > Average_Grade_Received
not_nan = df[df['Average_Grade_Received'].notna()]
difference_gpa = not_nan['Average_Grade_Received'] - not_nan['Average_Grade_Expected'].dropna()
difference_gpa.head

#Conclude that student will overestimate the grade recieved compared to actual grade they recieve!
#Actual grade they recieve are better estimate than estimted grade!
```

```python
#On average, Average_Grade_Expected are 0.20 GPA higher than Average_Grade_Received
difference_gpa.mean()
```

```python
#And this distribution is not that spread out - reference the histogram below, which is good!
difference_gpa.std()
```

```python
difference_gpa.plot.hist();
plt.xlabel('Difference (Average_Grade_Received - Average_Grade_Expected)')
plt.ylabel('Frequency')
plt.title('Difference in Grade Receieve and Grade Expected')
```

This plot shows that the difference in average grade received was for the most part in allignment with average grade expected.
```python
#Imputed Average_Grade_Received with Average_Grade_Expected with the average difference in GPA!
df['Average_Grade_Received'] = df['Average_Grade_Received'].fillna(df['Average_Grade_Expected'] + difference_gpa.mean() )
df['Average_Grade_Received'] = df['Average_Grade_Received'].round(2)
df
```

# Hypothesis Testing & Model

## Hypothesis Testing

Null Hypothesis: We hypothesize that the GPA you get for taking a certain course is not affected by which quarter you are taking the course
<br>
Alternative Hypothesis: We hypothesize that GPA you get for taking a certain courses is affected by which quarter you are enrolled

### Prepping the Dataset

Our hypothesis states that the GPA you get for taking a certain course is not affected by which quarter you are taking the course. It is important to know that besides the regular fall, winter, and spring quarter, our data also includes all summer quarters as well, which only last for half the time as a normal quarter. Looking at the average GPA recieved by quarter each year graph in the EDA section, we see that the average GPA heavily fluctuates during summer quarters. This may be because the time span of a summer quarter makes some courses exponentially more difficult, drastically decreasing the GPA, or that students may be more motivated if they are enrolling in summer classes than the regular student, which could be why GPA could be drastically increased. Another reason is that in summer session quarters, only certain classes are offered, which could create bias in our analysis. This is especially relevant in summer session 3, where the average summer session 3 quarter GPA is extremely high in every year except 2018. This is because the least number of classes are offered in summer session 3, which spikes up the GPA because of this. In any case, including summer quarter GPA data could definitely affect the analysis of our hypothesis, making this a potential confound. Therefore we have elected to remove summer quarter data, shown in the code block below.

```python
capes_without_summer = capes_df[~capes_df["Quarter"].isin(['S1','S2','S3','SU'])]
capes_without_summer
```

After removing the summer quarter rows, we wanted to replot the average GPA received by quarter each year graph to see what has changed.

```python
capes_without_summer_mean = capes_without_summer.groupby(["Year","Quarter"])["Average_Grade_Received"].mean()
years, quarters, values = zip(*[(year, quarter, value) for (year, quarter), value in capes_without_summer_mean.items()])

# Plotting
plt.figure(figsize=(16, 8))
plt.plot(range(len(values)), values, color = 'black', marker = 'o', markerfacecolor = 'red')
plt.xticks(range(len(values)), ['{} {}'.format(year, quarter) for year, quarter in zip(years, quarters)], rotation=75)
plt.title('Average GPA Received by Quarter Each Year')
plt.xlabel('Year & Quarter')
plt.ylabel('Average GPA')
plt.show()
```

Comparing this graph to the same graph in the EDA section, this graph is a lot less spiky because we removed the summer session 3 instances, which inflated the average GPA significantly. We still see the upward GPA trend over time, which is relatively stable from 2007-2018 until COVID hit from 2019-2021, which drastically increased the average GPA, peaking at over 3.6 at fall 2020. From 2020-2023, the average GPA still heavily fluctuates but is still significantly higher than pre-COVID times.

### T Test

Now we will perform hypothesis testing through t-tests. We decided to perform 3 t-tests comparing each combination of the three quarters (fall vs. winter, fall vs. spring, winter vs. spring) for each year. We decided to only consider years 2008-2022 because these are the years with all three quarters that have an average GPA datapoint. This means that we will be doing a total of 45 t-tests. We will tally up how many times we accept and reject the null hypothesis and observe the results.

```python
quarters = capes_without_summer['Quarter'].unique()
years = range(2008,2023)
tally = {'reject': 0, 'fail to reject': 0}
alpha = 0.05


for year in years:
    # print(f"Year: {year}")
    for i in range(len(quarters)):
        for j in range(i+1, len(quarters)):
            quarter1 = quarters[i]
            quarter2 = quarters[j]
            gpa_quarter1 = capes_without_summer[(capes_without_summer['Quarter'] == quarter1) & (capes_without_summer['Year'] == year)]['Average_Grade_Received']
            gpa_quarter2 = capes_without_summer[(capes_without_summer['Quarter'] == quarter2) & (capes_without_summer['Year'] == year)]['Average_Grade_Received']
            t_statistic, p_value = ttest_ind(gpa_quarter1, gpa_quarter2, equal_var=False)
            # print(f"T-test between {quarter1} and {quarter2}:")
            # print(f"    t-statistic: {t_statistic}")
            # print(f"    p-value: {p_value}")
            if p_value < alpha:
                tally["reject"] += 1
                # print("    Reject the null hypothesis.")
            else:
                tally["fail to reject"] += 1

                # print("    Fail to reject the null hypothesis.")
```

Here is the <a href = "\T_test_result_full"> Link </a> to the full T-test result

### Conclusion for T-Test

This is the tally of how many t test rejected null and fail to reject null

```python
print(tally)
```

Looking at the tally dictionary we made, it seems that the majority of comparisons fail to reject the null hypothesis (38-7). The majority of the 7 are due to the COVID years. Drastic increases and decreases of average GPA throughout each quarter in these years cause extremely low p-values, which in turn reject the null hypothesis. Other times that reject the null hypothesis can be attributed to an irregular trend of average GPA over time. For example, in Fall 2018, there was a significantly lower average GPA compared to its adjacent quarters. Because of this, we can conclude that our hypothesis, which states that the GPA you get from a certain course is not affected by which quarter you take it, is true.

## Creating our Random Forest Regression Model

Prediction: We want to create a prediction model that predicts the GPA students expect for taking the course
        Some attribute that we are anticipating to improve the model are, which course was taken, professor income, professor rating, professor department, and the number of students enrolled in the class.


Supervised Machine Learning: training the model on input and output data
- Classification

We decided to use a Random Forest Model since it provides the highest accuracy given our test data input. A Random forest model generates many trees that are not mutually correlated, and this contributes to results that are more reliable and ultimately more accurate than simple decsion trees. To test validity we create an 80 20 split for our data. 80% of our data will be used for training and 20% will be used for testing. Like the hypothesis testing, we filter the dataframe to only fall, winter, and spring since we only want to analyze the academic year.

```python
filtered_df = merged_df[merged_df['Quarter'].isin(['FA', 'WI', 'SP'])]

# Assuming 'merged_df' is your DataFrame and contains all your data
X = filtered_df[['Course', 'Department', 'Income', 'Percentage_Recommended_Professor', 'Total_Enrolled']]  # Features
y = filtered_df['Average_Grade_Received']  # Target variable

# Split the data - 80% for training, 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Example preprocessing for numerical and categorical data
numerical_cols = X_train.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = X_train.select_dtypes(include=['object', 'bool']).columns

# Creating transformers for preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols),
    ])

# Fit on the training data and transform both training and testing data
X_train_preprocessed = preprocessor.fit_transform(X_train)
X_test_preprocessed = preprocessor.transform(X_test)
```

```python
# Initialize and train your model
model = RandomForestRegressor(n_estimators=100, random_state=0, oob_score=True)
model.fit(X_train_preprocessed, y_train)
```

```python
# Predict on the test set
y_pred = model.predict(X_test_preprocessed)

# Calculate performance metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Out-of-Bag Score: {model.oob_score_}')
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

# Ethics & Privacy

With the CAPES data we proposed, there seems to be no terms of use issues with the data as CAPES was a voluntary survey offered to students. The vast majority of classes have a very high percentage of CAPES participation and we have an adequate amount of data to make reliable inferences. However, there may be potential biases with our data. Often times, professors would offer a small amount of extra credit or some other incentive if a certain percentage of the class completes CAPES. This could create a bias toward students who care more about their grade which could lead to an inaccurate representation of the entire student body taking that class. Also, since our data is collected from a voluntary survey, the types of students that have a very positive or negative experience with the class would be more incentivized to voluntarily fill out CAPES. In summary, because of the way the data is collected and external incentives that may manipulate the data, our data may not be representative of all students’ experiences of the class.

As for potential biases during and after our analysis, there could be potential problems with exposing sensitive data about professors. # Discussion and Conclusion


The research question aimed to understand the expected grade from taking a certain course at UCSD and whether this GPA is influenced by the quarter in which the course was taken. We wanted to create a prediction model based on factors such as professor ratings, income, and course details. We drew insights from previous research, including Flaherty's study on student ratings of instruction and the COGS 108 Winter 2021 Group 46 project. Our hypothesis suggested that the GPA received for a course would not be significantly affected by the quarter, assuming consistency in grading practices and course difficulty across quarters. This assumption came from the belief that professors within the same department would agree on grading standards regardless of the time of year. To investigate this hypothesis, we used the UCSD Capes Dataset and the UC Income Dataset to obtain essential information on course details and professor income. We also used various tools to utilize and interpret the data, such as web scraping, data preprocessing, and Random Forest Regression, a machine learning algorithm.


Our exploratory data analysis gave us insight into UCSD's academics. We observed the dominance of STEM departments. Our analysis of study hours and average grades across departments uncovered patterns suggesting varying academic difficulty. We used a bar graph to analyze the average GPA per quarter over the years. From this, we came to the conclusion that all of the quarters had roughly the same GPA, with the outlier being quarter S3, and this is most clearly seen in the years 2013–2016. The analysis of the average GPA per quarter over the years led to the conclusion that, aside from outliers such as quarter S3, the GPA remained relatively consistent across quarters, supporting our hypothesis.


From the random forest regression model and our analysis of the correlation between quarter and average grade received, we can see that there is not a strong enough correlation to justify a causal relationship between which course is taken, professor income, department, professor rating, student count, and the average grade received. The model we created to predict grades was ineffective because of a lack of correlation and its inability to handle seen and unseen data. While uneffective, we were still able to answer the question proposed above. Some key findings worth mentioning are the professor ratings and grade-received correlations. We saw a positive correlation, showing that with a higher grade received, more students were likely to fill out the form saying good things about the professor. This is in line with what our previous work has mentioned. We added to their work by incorporating professors income, which helped showcase the resource allocation imbalance for STEM vs. non-STEM-based classes. This, however, didn't result in a direct effect on the average grade received. One thing worth noting on this is that we showed how the courses with the lower average GPA were those from the STEM departments that have more resources allocated. While the resources aren't connected to grades, it just shows that those departments are harder to succeed in, which is true for a variety of reasons mentioned (i.e., grading policies, work load, material covered).


The results of our hypothesis testing through various t-tests showed that our null hypothesis is mostly true due to the overwhelming majority of p-values that fail to reject the null hypothesis. This is because, for most years, the fluctuation of the average GPA between each quarter varies quite little. There could be some outliers that have extremely different GPAs for the same class but in a different quarter each year due to different professors, styles of teaching, outside events, etc., but for the most part, on average, the GPA variation is not big at all. The majority of quarter comparisons that did reject the null hypothesis were due to extreme fluctuations in average GPA during 2020–2022, or when COVID happened. Of course, COVID is an external factor that affects a student's GPA. Professors let their guidelines become more lax during the pandemic, and every class became online, so it was easier to do well in classes during this time period. This was not the norm of how classes were supposed to operate, so we determined that the majority of comparisons that rejected the null hypothesis were invalid and that we could safely conclude that we do not have enough evidence to reject our null hypothesis.
